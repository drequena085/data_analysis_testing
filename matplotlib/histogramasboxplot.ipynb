Gráficos de Histograma y Boxplot para distribuciones.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(170, 10, 200)
print(data)
```

Los intervalos de información por defecto es de 10, con el parámetro "bins" lo podemos editar.

```python
plt.hist(data, color='salmon', bins=20, edgecolor='black', alpha=0.6)
plt.title('Distribución de alturas')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')


plt.show()
```

```python
np.random.seed(0)
ages = [np.random.normal(30, 5, 100), np.random.normal(40, 5, 100), np.random.normal(35, 5, 100)]
#print(ages)
```

```python
plt.boxplot(ages, patch_artist=True, notch=True, vert=True, tick_labels=['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.title('Distribución de edades por grupo')
plt.xlabel('Grupo')
plt.ylabel('Edad (años)')


plt.show()
```
