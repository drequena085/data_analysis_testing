```python
import matplotlib.pyplot as plt

hours = [2,3,4,5,6,7,8]
exam_scores_s1 = [55,60,65,70,75,80,85]
exam_scores_s2 = [50,58,63,69,74,78,83]

# Gráfico de disperción de dos estudiantes
plt.scatter(hours, exam_scores_s1, marker='o', color='green', label='Estudiante 1')
plt.scatter(hours, exam_scores_s2, marker='s', color='red', label='Estudiante 2')
plt.title('Relación entre horas estudiadas y el puntaje')
plt.xlabel('Horas')
plt.ylabel('Puntaje')
plt.show()
```

```python
plt.plot(hours, exam_scores_s1, marker='o', color='green', linestyle='-', linewidth=2, label='Estudiante 1')
plt.plot(hours, exam_scores_s2, marker='s', color='red', linestyle='--', linewidth=2, label='Estudiante 2')
plt.title('Relación entre horas estudiadas y el puntaje')
plt.xlabel('Horas')
plt.ylabel('Puntaje')
plt.show()
```
