```python
import numpy as np

data = np.array([[3,2,2],[4,5,6],[7,8,9]])

df_from_array = pd.DataFrame(data, columns=['A','B','C'])
print(df_from_array)
```

```python
data = [[1, 'Jhon', 22], [2,'Anna', 24]]

df_from_list = pd.DataFrame(data, columns=['Id','Name', 'Age'])
print(df_from_list)
```

```python
data = [{'ID': 1,
         'Name': 'Duvan',
         'Age': 25},
         {'ID': 2,
          'Name': 'Andry',
          'Age': 22}]

df_from_dict = pd.DataFrame(data)
print(df_from_dict)
```

```python
data = {'ID': [1,2,3], 'Name': ['John', 'Anna', 'Peter'], 'Age': [22,24,21]}
df_from_dict2 = pd.DataFrame(data)
print(df_from_dict2)
```

```python
data = {'ID': pd.Series([1,2,3]),
         'Name': pd.Series(['John', 'Anna', 'Peter']),
         'Age': pd.Series([22,24,25])}

df_from_series_dict = pd.DataFrame(data)
print(df_from_series_dict)
```
