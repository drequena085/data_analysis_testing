Datos faltantes

```python
import pandas as pd

path = './online_retail.csv'
retail_data = pd.read_csv(path)
```

usando la función ".isna()" podemos descubrir en qué columna tenemos datos faltantes. Si el dato no está disponible dirá True

```python
missing_data = retail_data.isna()
print(missing_data)
```

Así podemos contar cuantos registros cuentan con datos faltantes y verlo por columna.

```python
# print(missing_data.head())
missing_data_count = retail_data.isna().sum()
print('Conteo de datos faltantes por columna: \n ', missing_data_count)
```

Así podemos eliminar las filas con los datos faltantes

```python
no_missing_rows = retail_data.dropna()
print('Datos sin filas con valores faltantes \n', no_missing_rows )
```

axis por defecto es 0, que corresponde a las filas, 1 a las columnas. Así borramos las columnas con datos faltantes

```python
no_missing_columns = retail_data.dropna(axis=1)
print(no_missing_columns)
```

```python
retail_data_fill_zeros = retail_data.fillna(0)
# print(retail_data_fill_zeros)
missing_data_count = retail_data_fill_zeros.isna().sum()
print(missing_data_count)
```

```python
mean_unit_price = retail_data['UnitPrice'].mean()
retail_data_filled_mean = retail_data['UnitPrice'].fillna(mean_unit_price)
print(retail_data_filled_mean)
```
