Groupby

```python
import pandas as pd

path = './online_retail.csv'
df = pd.read_csv(path)
```

```python
country_count = df['Country'].value_counts()
print(country_count)
```

```python
country_group = df.groupby('Country')['Quantity'].sum()
print(country_group)
```

```python
country_stats = df.groupby('Country')['UnitPrice'].agg(['mean', 'sum'])
print(country_stats)
```

```python
country_stock_group = df.groupby(['Country', 'StockCode'])['Quantity'].sum()
print(country_stock_group)

registro = df[ (df['StockCode'] == '15036') & (df['Country'] == 'Australia')]
# print(registro)
```

```python
def total_revenue(group):
    return (group['Quantity'] * group['UnitPrice'].sum())

revenue_per_country = df.groupby('Country').apply(total_revenue)
print(revenue_per_country)
```

Ejercicio: Top 3 mejores y peores países en ventas

```python

df['total_price'] = df['Quantity'] * df['UnitPrice']

ventas_pais = df.groupby('Country')['total_price'].sum()
ventas_pais = ventas_pais.reset_index()
ventas_pais = ventas_pais.sort_values(by='total_price', ascending=False)
print('Top 3: \n ', ventas_pais.head(3))
print('Bottom 3: \n ', ventas_pais.tail(3))
```
