Estructuras y funciones

```python
import pandas as pd

path = './online_retail.csv'
retail_data = pd.read_csv(path)

column_names = retail_data.columns
print(column_names)
```

```python
num_row, num_columns = retail_data.shape
print(num_row)
print(num_columns)
```

```python
daily_sales = retail_data['Quantity']
print(daily_sales[2])
```

```python
summary = retail_data.describe()
print(summary)
```

```python
mean_value = daily_sales.mean()
print(mean_value)
median_value = daily_sales.median()
print(median_value)
tota_sum = daily_sales.sum()
print(tota_sum)
```

```python
count_values = daily_sales.count()
print(count_values)
```
