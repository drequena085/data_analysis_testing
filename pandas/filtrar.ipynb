Filtrar ventas por país

```python
import pandas as pd

path = './online_retail.csv'
df = pd.read_csv(path)

df['total_price'] = df['Quantity'] * df['UnitPrice']

uk_sales = df[ df['Country'] == 'United Kingdom']
print(uk_sales)
```

```python
high_quantity_sales = df[ df['Quantity'] > 200 ]
print(high_quantity_sales)
```

```python
uk_high_sales = df[(df['Quantity'] > 1000) & (df['Country'] == 'United Kingdom')]
print(uk_high_sales)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
sales_2010 = df[ df['InvoiceDate'].dt.year == 2010 ]
print(sales_2010)
```

```python
sales_dec_2010 = df[ (df['InvoiceDate'].dt.year == 2010) & df['InvoiceDate'].dt.month == 12]
print(sales_2010)
```
