Cálculos Matriciales

```python
import numpy as np

#SUMA DE MATRICES

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

sum = A + B
print(sum)
```

```python
#MULTIPLICACION DE MATRICES

product = np.dot(A, B)
print(product)
```

```python
#Determinante e inversa

det_A = np.linalg.det(A)
print(det_A)

inv_A = np.linalg.inv(A)
print(inv_A)
```

```python
#Resolver AX = b

b = np.array([9, 11])
x = np.linalg.solve(A, b)
print(x)
```
