```python
import numpy as np

# Supongamos que tenemos la variable 'respuestas' con las opiniones de los clientes
respuestas = np.array(['bueno', 'excelente', 'malo', 'bueno', 'excelente', 'bueno', 'malo', 'excelente'])

print(np.unique(respuestas))
```

```python
unique_elements, counts = np.unique(respuestas, return_counts = True)
print(unique_elements)
print(counts)
```

Importante tener en cuenta que view_y apunta a la misma dirección de memoria que la lista original, así que si la lista original se modifica, también lo hará el view.

```python
array_x = np.arange(10)
view_y = array_x[1:3]
print(array_x)
print(view_y)

array_x[1:3] = [10,11]

print(array_x)
print(view_y)
```

Cuando hacemos doble llaves [[]] podemos copiar el array sin apuntar a la misma dirección de memoria

```python
copy_x = array_x[[1,2]]

array_x[1:3] = [14,15]
print(array_x)
print(copy_x)
```
