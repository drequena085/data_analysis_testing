Arrays en Numpy

.ndim -> Para ver las dimensiones.
.shape -> Para ver la medida de cada dimensión.
.dtype -> el datatype de los registros.

```python
import numpy as np

array = np.array([[1,2,3],[4,5,6]])
print(array.ndim)
print(array.shape)
print(array.dtype)
```

Datatype: Es un objeto de numpy que describe el tipo de elementos dentro del array, puede ser entero, flotante o booleano, también el tamaño del dato, ejemplo: 8bits, 16bits, etc.  

uint8 -> Es un entero sin signo de 8bits, puede representar valores entre 0 y 255. útil para trabajar con datos que no pueden ser negativos como colores donde cada color puede tener un valor entre 0 y 255.

float32 -> número de punto flotante de 32bits. Representa valores con decimales y se usan para manejar precisión moderada pero donde se requiere ahorra bits.

float64 -> Lo mismo que el anterior pero de 64bits. Ofrece mayor precisión y es el valor por defecto para decimales que usa numpy.

```python
z = np.array(3, dtype=np.uint8)
print(z)
```

dtype='d' los convierte a número flotante de 64bit por defecto

```python
double_array = np.array([1,2,3], dtype='d')
print(double_array)
```

```python
z = z.astype(np.float64)
print(z)
```

Operaciones matemáticas entre matrices:

SUMA DE MATRICES:

```python
A = np.array([[1, 2], [3, 4],])
B = np.array([[5, 6], [7, 8]])

suma = A + B
print("Suma de matrices:\n", suma)
```

PRODUCTO DE MATRICES

```python
producto = np.dot(A, B)
print("Producto de matrices:\n", producto)
```

DETERMINANTE DE UNA MATRIZ (SOLO VALIDO EN UNA MATRIZ CUADRADA)

```python
determinante = np.linalg.det(A)
print("Determinante de A:", determinante)
```

INVERSA DE UNA MATRIZ

```python
inversa = np.linalg.inv(A)
print("Inversa de A:\n", inversa)
```

Valores y vectores propios

Los valores propios y los vectores propios son fundamentales en muchas aplicaciones, como en la compresión de datos y el análisis de sistemas dinámicos.

```python
valores_propios, vectores_propios = np.linalg.eig(A)
print("Valores propios de A:\n", valores_propios)
print("Vectores propios de A:\n", vectores_propios)
```

Resolución de sistemas de ecuaciones lineales

Para resolver un sistema de ecuaciones lineales AX=BAX = BAX=B:

```python
B = np.array([1, 2])
X = np.linalg.solve(A, B)
print("Solución del sistema AX = B:\n", X)
```
