```python
import numpy as np


meses = np.array(['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                  'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'])

ventas_A = np.array([150, 200, 250, 300, 220, 210, 180, 190, 230, 240, 280, 300])
ventas_B = np.array([180, 210, 230, 250, 270, 260, 240, 250, 270, 290, 310, 330])
ventas_C = np.array([200, 220, 240, 260, 280, 300, 320, 340, 360, 380, 400, 420])

print("Media de ventas Producto A: ", np.mean(ventas_A))
print("--> Suma de ventas Producto A: ", np.sum(ventas_A))
print("Media de ventas Producto B: ", np.mean(ventas_B))
print("--> Suma de ventas Producto B: ", np.sum(ventas_B))
print("Media de ventas Producto C: ", np.mean(ventas_C))
print("--> Suma de ventas Producto C: ", np.sum(ventas_C))

suma = ventas_A + ventas_B + ventas_C
result = list(zip(meses, suma))
for month, total in result:
    print('[' + str(month) + '] Total: ' + str(total))
```

```python
mes_mayor_venta = meses[np.argmax(suma)]
mes_menor_venta = meses[np.argmin(suma)]
print(f"Mes con mayor ventas: {mes_mayor_venta}")
print(f"Mes con menor ventas: {mes_menor_venta}")
```
